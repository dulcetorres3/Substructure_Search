# Substructure Screen

Acetaminophen is the active ingredient in the popular medication, Tylenol. It consists of several functional groups including amide, hydroxyl, and aromatic groups. We will first take a look at what our target compound looks like.

```python
from molecule import MoleculeGraph

acetaminophen =  MoleculeScreen.from_sdf('acetaminophen.sdf', True)
acetaminophen.visualization
```

As we can see, there are no double or triple bonds portrayed in the visalization. However, the bond type is recorded within the attribute of each bond. We will now take a look at structures we will be screening for within the compound. 

```python
benzene =  MoleculeScreen.from_sdf('benzene.sdf', True)
benzene.visualization
```

```python
hydroxyl =  MoleculeScreen.from_sdf('hydroxy.sdf', True)
hydroxyl.visualization
```

```python
cyclohexane =  MoleculeScreen.from_sdf('cyclohexane.sdf', True)
cyclohexane.visualization
```

```python
nitro =  MoleculeScreen.from_sdf('nitro.sdf', True)
nitro.visualization
```

We will now screen using the structures above: benzene, hydroxyl, cyclohexane, and nitro. Although there is resemblance between benzene and cyclohexane, we expect a negative hit for the latter due to the excess hydrogens. Similarly, although both amide and nitro groups share a nitrogen, we expect a negative hit for N02. Therefore, bezene and hydroxyl are the only structures that we expect a postivie hit for.

```python
print(f"The benzene group is a substructure of acetaminophen: {acetaminophen.screen(benzene)}")
print(f"The hydroxyl group is a substructure of acetaminophen: {acetaminophen.screen(hydroxyl)}")
print(f"The cyclohexane group is a substructure of acetaminophen: {acetaminophen.screen(cyclohexane)}")
print(f"The nitro group is a substructure of acetaminophen: {acetaminophen.screen(nitro)}")
```
